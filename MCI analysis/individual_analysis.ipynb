```python
root_folder = "./MCI analysis"
# subject = 'mci008'  # List of subjects
subjects = ['mci001', 'mci002', 'mci003', 'mci004', 'mci005', 'mci006', 'mci007', 'mci008', 'mci009', 'mci010', 'mci014']  # List of subjects
files = ['mci001_right_df.csv', 
            'mci002_left_df.csv', 
            'mci003_right_df.csv', 
            'mci004_left_df.csv', 
            'mci005_left_df.csv', 
            'mci006_right_df.csv', 
            'mci007_left_df.csv', 
            'mci008_left_df.csv', 
            'mci009_left_df.csv', 
            'mci010_right_df.csv', 
            # 'mci011_left_df.csv', 
            # 'mci012_left_df.csv', 
            # 'mci013_left_df.csv',
            # 'mci014_left_df.csv',
            # 'mci016_left_df.csv',
            # 'mci017_left_df.csv'
         ]


for file in files:
    dataFrames = {}
    # files = [file for file in root_files if any(subject in file for subject in subjects) and 'right' not in file]
    # file = f'{subject}_left_df.csv'
    subject = file.split('_')[0]

    file_path = os.path.join(root_folder, 'data', 'compiled_data', file)
    df = pd.read_csv(file_path)

    df = df.drop(['Session', 'Unnamed: 0'], axis=1)
    df['Date'] = pd.to_datetime(df['Date'], format='%Y-%m-%d')

    print(df.columns)

    aggregation_functions = {
        col: 'mean' for col in df.columns if col != 'Total_Activity_Time_minutes'
    }
    aggregation_functions['Total_Activity_Time_minutes'] = 'sum'

    # Group by 'Date' and apply aggregation functions
    grouped_df = df.groupby('Date').agg(aggregation_functions)

    grouped_df = grouped_df.drop(['Date', 'Avg_Gait_Speed', 'Avg_COP_Speed'], axis=1)
    print(grouped_df.columns)
    grouped_df.columns = ['total_time', 'step_time', 'swing_time', 'stride_time', 'total_steps', 'cadence', 'step_variability', 'stride_variability', 'median_cop_speed', 'total_activity']
    grouped_df.index = grouped_df.index.strftime('%Y-%m-%d')
    grouped_df.to_json(f'{subject}.json')
```
